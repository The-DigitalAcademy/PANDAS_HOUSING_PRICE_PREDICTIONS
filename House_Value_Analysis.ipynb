```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
num_var = ['longitude',	'latitude',	'housing_median_age',	'total_rooms','total_bedrooms',	'population',	'households',	'median_income',	'median_house_value']

plt.figure(figsize=(25,25))
# Calculate the number of rows and columns for the subplots
num_rows = (len(num_var) + 3) // 4 
num_cols = 4
for i, var in enumerate(num_var):
    plt.subplot(num_rows, num_cols, i + 1)
    sns.distplot(df[var], bins=20)
    plt.title(var)  

plt.tight_layout()
plt.show()
```

```python


# Calculate the average house value for each ocean proximity category
avg_house_values = df.groupby('ocean_proximity')['median_house_value'].mean()

# Calculate bubble size based on the average house value (you might need to adjust the scaling factor)
bubble_sizes = avg_house_values / 1000  # Adjust the scaling factor for appropriate bubble sizes

# Plotting the bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(
    avg_house_values.index,
    avg_house_values,
    s=bubble_sizes,
    alpha=0.7,
    c=avg_house_values,
    cmap='viridis'
)

# Adding labels and title
plt.xlabel('Ocean Proximity')
plt.ylabel('Average House Value')
plt.title('Average Housing Values by Ocean Proximity')
plt.colorbar(label='Average House Value')

# Adjust layout
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```

```python

avg_values = df.groupby('ocean_proximity')[['median_house_value', 'housing_median_age']].mean()

ocean_proximities = avg_values.index
avg_house_values = avg_values['median_house_value']
avg_house_age = avg_values['housing_median_age']

# Creating the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(
    avg_house_age,
    avg_house_values,
    s=100,  # Marker size
    c='pink',  # Marker color
    alpha=0.7
)

# Adding labels and title
plt.xlabel('Median House Age')
plt.ylabel('Average House Value')
plt.title('Ocean Proximity Impact on Housing Value and Median House Age')

# Adding data labels for each point
for i, prox in enumerate(ocean_proximities):
    plt.annotate(prox, (avg_house_age[i], avg_house_values[i]), textcoords="offset points", xytext=(0, 10), ha='center')

# Show the plot
plt.grid(False)
plt.tight_layout()
plt.show()
```

```python

avg_values = df.groupby('ocean_proximity')[['median_house_value', 'total_bedrooms']].mean()

ocean_proximities = avg_values.index
avg_house_values = avg_values['median_house_value']
avg_total_bedrooms = avg_values['total_bedrooms']

plt.figure(figsize=(10, 6))
plt.scatter(
    avg_total_bedrooms,
    avg_house_values,
    s=100,  # Marker size
    c='purple',  # Marker color
    alpha=0.7
)


plt.xlabel('Average Total Bedrooms')
plt.ylabel('Average House Value')
plt.title('Analysis of Housing')

for i, prox in enumerate(ocean_proximities):
    plt.annotate(prox, (avg_total_bedrooms[i], avg_house_values[i]), textcoords="offset points", xytext=(0, 10), ha='center')

plt.grid(False)
plt.tight_layout()
plt.show()
```

```python

sns.set(style="whitegrid")

fig, ax1 = plt.subplots(figsize=(10, 6))

sns.barplot(x='ocean_proximity', y='median_house_value', data=df, ax=ax1, color='turquoise')

ax1.set_xlabel('Ocean Proximity')
ax1.set_ylabel('Median House Value', color='turquoise')
ax1.tick_params(axis='y', labelcolor='turquoise')
ax1.set_title('House Value and Income by Ocean Proximity')

ax2 = ax1.twinx()

sns.lineplot(x='ocean_proximity', y='median_income', data=df, ax=ax2, color='lightpink')

ax2.set_ylabel('Median Income', color='lightpink')
ax2.tick_params(axis='y', labelcolor='lightpink')

plt.tight_layout()

plt.xticks(rotation=45)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

fig, ax1 = plt.subplots(figsize=(10, 6))

# Sorting DataFrame by median_house_value in descending order
df_sorted = df.sort_values(by='median_house_value', ascending=False)

sns.barplot(x='ocean_proximity', y='median_house_value', data=df_sorted, ax=ax1, color='turquoise', linewidth=0)

ax1.set_xlabel('Ocean Proximity')
ax1.set_ylabel('Median House Value', color='turquoise')
ax1.tick_params(axis='y', labelcolor='turquoise')
ax1.set_title('House Value and Income by Ocean Proximity')

ax2 = ax1.twinx()

sns.lineplot(x='ocean_proximity', y='median_income', data=df_sorted, ax=ax2, color='lightpink')

ax2.set_ylabel('Median Income', color='lightpink')
ax2.tick_params(axis='y', labelcolor='lightpink')

ax1.grid(False)  # Removing grid lines from the first axis
ax2.grid(False)  # Removing grid lines from the second axis

plt.tight_layout()

plt.xticks(rotation=45)
plt.show()
```
